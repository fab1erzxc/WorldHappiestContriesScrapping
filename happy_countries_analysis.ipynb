```python
from bs4 import BeautifulSoup as bs
import requests
```

```python
url = r'https://en.wikipedia.org/wiki/World_Happiness_Report'

r = requests.get(url)

soup = bs(r.text, 'html.parser')
```

```python
soup
```

```python
headers = soup.find_all('th')[1:3]
headers = [header.text.strip() for header in headers]
print(headers)
```

```python
import pandas as pd
```

```python
df = pd.DataFrame(columns=headers)
df
```

```python
columns = soup.find_all('tr')
columns = [column.text.strip() for column in columns][3:140]
columns
```

```python
columns = [column.split('\n') for column in columns]
columns = [[int(i), j[1:]] for i, j in columns]
columns
```

```python
ranks = [i[0] for i in columns]
ranks
```

```python
df["Overall rank"] = ranks
df
```

```python
countries = [i[1] for i in columns]
countries
```

```python
df["Country or region"] = countries
df
```

```python
df.to_csv('world_happiness_report.csv', index=False)
```
